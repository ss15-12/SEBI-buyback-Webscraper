```python
from selenium.webdriver.common.by import By
main=[]
for i in range(1,10):
    driver.get('https://www.chittorgarh.com/buyback/natco-pharma-buyback-2023/'+str(i)+'/')
    table_trs = driver.find_elements(By.XPATH, "//table[contains(@class,'table table-bordered table-striped w-auto')]")
    value_list = []
    for v in range(len(table_trs )):
        value_list.append(table_trs [v].text)
#     for i in value_list:
#         print(i)
#     print('\n')
    main.append(value_list)
```

```python
import pandas as pd 
#COLUMN_NAMES=['ISSUE size','Security name','Buyback price','Face value','listing at']
df = pd.DataFrame()
for i in main:
    issue_lines = i[0].split('\n')
    print(issue_lines)
    issue_details = {}
    for line in issue_lines:
        if line:
            #key, value = line.split(' ', 1)
            key = ' '.join(line.split()[:2])
            value = ' '.join(line.split()[2:])
            issue_details[key] = value
    df=df.append([issue_details])
```

```python
from selenium.webdriver.common.by import By
main2=[]
for i in range(1,10):
    driver.get('https://www.chittorgarh.com/buyback/natco-pharma-buyback-2023/'+str(i)+'/')
    table_trs = driver.find_elements(By.XPATH, "/html/body/div[7]/div[3]/div[1]/div[6]/div[2]/div/table")
    value_list2 = []
    for v in range(len(table_trs )):
        value_list2.append(table_trs [v].text)
#     for i in value_list:
#         print(i)
#     print('\n')
    main2.append(value_list2)
```

```python
df2=pd.DataFrame()
for i in main2:
    if i:
        for j in i:
            lines = j.split('\n')
            data = {}
            for line in lines:
                if line:
                    key = ' '.join(line.split()[:-3])  # Assign everything except the last two elements as the key
                    date = line.split()[-3:]  # Extract the last two elements of the line as the month and day
                    date_str = ' '.join(date)
                    data[key] = date_str
            df2=df2.append([data])
    else:
        s2 = pd.Series([0,0,0,0,0,0,0], index=['Record Date', 'Offer Opens On', 'Last Date for receipt of Tender Forms', 'Offer Closes On *','Finalisation of Buyback Acceptance','Last Date for settlment of bids','Last Date for Extinguishment of Shares'])
        df2 = df2.append(s2,ignore_index=True)
```
